```python
shirt_count=df["name"].head().value_counts()
shirt_count.plot(kind="pie",autopct="%.2f%%")
plt.title("shirt type popularity")
plt.show()
```

```python
ori_price_count=df["ori_price"].head().value_counts()
ori_price_count.plot(kind="pie",autopct="%.2f%%")
plt.title("mostly buyed shirts at what price")
plt.show()
```

```python
dis_price_count=df["discount"].tail().value_counts()
dis_price_count.plot(kind="pie",autopct="%.2f%%")
plt.title("discount given")
plt.show()
```

```python
df.discount=df.discount.str.replace('% off','')
```

```python
df.discount=df.discount.astype(int)
```

```python
df["discount"].plot.density()
```

```python
df.dis_price.head()
```

```python
df.dis_price=df.dis_price.str.replace("₹",'')
```

```python
df.dis_price=df.dis_price.astype(int)
```

```python
df["dis_price"].plot.density()
```

```python
df.ori_price=df.ori_price.str.replace("₹",'')
```

```python
df.ori_price=df.ori_price.str.replace(",",'')
```

```python
df.ori_price=df.ori_price.astype(int)
```

```python
plt.style.use("seaborn-poster")
plt.barh(df["discount"].head(20),df["ori_price"].head(20),label="discount")
plt.legend()
plt.title("discount vs ori_price")
```

```python
data=pd.DataFrame(df,columns=['ori_price','dis_price','discount'])
data.hist()
plt.show()
```
